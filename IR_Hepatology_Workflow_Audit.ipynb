# Clinical Background

## End-Stage Liver Disease and Fluid Accumulation

Patients with end-stage liver disease (ESLD) and cirrhosis commonly develop complications related to portal hypertension and hepatic dysfunction. Among the most frequent manifestations are ascites and hepatic hydrothorax, both of which can significantly reduce quality of life and contribute to repeated hospital admissions.

Ascites refers to pathological accumulation of fluid within the peritoneal cavity, while hepatic hydrothorax refers to pleural fluid accumulation occurring in the setting of portal hypertension. Patients often present with:
- Abdominal distension
- Dyspnea
- Pain and discomfort
- Reduced mobility
- Respiratory compromise

These conditions frequently require therapeutic fluid drainage procedures such as therapeutic paracentesis and thoracentesis for symptomatic relief.

---

## Therapeutic Paracentesis and Thoracentesis

Therapeutic paracentesis and thoracentesis are commonly performed procedures in hepatology practice and are generally considered safe when performed under appropriate procedural guidance and monitoring.

However, many patients with advanced liver disease also present with:
- Elevated INR
- Thrombocytopenia
- Impaired coagulation profiles

This creates concern regarding procedural bleeding risk and often results in additional coagulation assessment or blood product correction prior to intervention. Coagulation correction strategies may vary between ROTEM-guided transfusion and clinician-directed blood product administration based on conventional coagulation parameters and clinical assessment.

ROTEM-guided coagulation assessment has increasingly been explored as a strategy for targeted blood product utilization in patients with liver disease and coagulopathy.

# Cleaning and Standardizing Column Names

Column names were simplified and standardized to improve readability and facilitate workflow interval analysis.

```python
pros_df.columns = [
    'SNo',
    'Date',
    'Location',
    'Procedure',
    'Indication',
    'Hepatology_Review_Time',
    'Coagulopathy',
    'Transfusion_Strategy',
    'ROTEM_Request_Time',
    'ROTEM_Report_Time',
    'ROTEM_Result',
    'Number_of_Products',
    'FFP',
    'CRYO',
    'Platelet',
    'Blood_Product_Ordering_Time',
    'Blood_Product_Receiving_Time',
    'Transfusion_Start_Time',
    'Transfusion_End_Time',
    'IR_Inform_With_Correction',
    'IR_Inform_Without_Correction',
    'Shift_To_Procedure',
    'Procedure_Start_Time',
    'Shift_To_Recovery',
    'Shifted_To_Ward_Time',
    'Post_Procedure_Complications',
    'Other_Complications'
]
```

```python
pros_df.head()
```

# Standardizing Categorical Variables

Categorical variables were standardized to remove formatting inconsistencies and duplicate category representations.

```python
pros_df['Coagulopathy'] = (
    pros_df['Coagulopathy']
    .astype(str)
    .str.strip()
    .str.upper()
)

pros_df['Location'] = (
    pros_df['Location']
    .astype(str)
    .str.strip()
    .str.upper()
)
```

```python
strategy_map = {
    1: 'ROTEM-guided',
    2: 'Clinician-directed'
}

pros_df['Transfusion_Strategy'] = (
    pros_df['Transfusion_Strategy']
    .map(strategy_map)
)
```

# Converting Workflow Timestamps

Workflow timestamps were converted into datetime format to enable calculation of workflow intervals and procedural delays.

```python
time_columns = [
    'Hepatology_Review_Time',
    'ROTEM_Request_Time',
    'ROTEM_Report_Time',
    'Blood_Product_Ordering_Time',
    'Blood_Product_Receiving_Time',
    'Transfusion_Start_Time',
    'Transfusion_End_Time',
    'IR_Inform_With_Correction',
    'IR_Inform_Without_Correction',
    'Shift_To_Procedure',
    'Procedure_Start_Time',
    'Shift_To_Recovery',
    'Shifted_To_Ward_Time'
]

for col in time_columns:
    pros_df[col] = pd.to_datetime(
        pros_df[col],
        format='%H:%M',
        errors='coerce'
    )
```

# Workflow Delay Parameter Creation

Workflow intervals were calculated prospectively to identify procedural bottlenecks across different stages of the workflow pathway.

```python
pros_df['ROTEM_Delay_Min'] = (
    pros_df['ROTEM_Report_Time'] -
    pros_df['ROTEM_Request_Time']
).dt.total_seconds() / 60

pros_df['Blood_Product_Processing_Delay_Min'] = (
    pros_df['Blood_Product_Receiving_Time'] -
    pros_df['Blood_Product_Ordering_Time']
).dt.total_seconds() / 60

pros_df['Transfusion_Duration_Min'] = (
    pros_df['Transfusion_End_Time'] -
    pros_df['Transfusion_Start_Time']
).dt.total_seconds() / 60

pros_df['Procedure_Shift_Delay_Min'] = (
    pros_df['Procedure_Start_Time'] -
    pros_df['Shift_To_Procedure']
).dt.total_seconds() / 60

pros_df['Recovery_To_Ward_Delay_Min'] = (
    pros_df['Shifted_To_Ward_Time'] -
    pros_df['Shift_To_Recovery']
).dt.total_seconds() / 60

pros_df['Total_Workflow_Duration_Min'] = (
    pros_df['Procedure_Start_Time'] -
    pros_df['Hepatology_Review_Time']
).dt.total_seconds() / 60
```

- Parallel initiation of blood bank coordination and procedural planning where clinically appropriate
- Earlier hepatology assessment for anticipated procedures
- Standardized coagulation correction criteria
- Parallel workflow processing where feasible
- Earlier blood product preparation and communication with blood bank services
- Structured interdepartmental communication pathways
- Priority procedural coordination for high-risk symptomatic patients
- Standardized workflow timestamp documentation

Implementation of these measures may improve procedural efficiency, reduce avoidable delays, and improve multidisciplinary workflow coordination.

The proposed workflow pathway was designed in alignment with healthcare quality improvement principles and existing hepatology procedural safety practices.

# Limitations


Several limitations were identified during this audit.

The retrospective component was limited to a one-month review period and was additionally affected by inconsistent workflow timestamp documentation.

Several retrospective workflow intervals could not be accurately evaluated due to incomplete procedural recording practices.

Additional workflow variability was introduced by operational factors including variation in hepatology review timing, IR operating hours, blood product availability, and dependency on sequential multidisciplinary coordination.

Workflow progression was dependent on hepatology review and coagulation correction decisions prior to IR  referral.

The prospective component, although more structured, was conducted within a single workflow environment.

# Future Work

Future prospective evaluation over longer study durations is recommended to:

- Validate workflow findings
- Assess the impact of workflow optimization strategies
- Evaluate reduction in procedural delays
- Improve workflow standardization
- Develop institution-specific procedural fast-track pathways

# Clinical and Service Impact

Improving workflow efficiency for therapeutic paracentesis and thoracentesis may contribute to:

- Earlier symptom relief
- Improved patient flow
- Reduced procedural waiting times
- Better multidisciplinary coordination
- Improved utilization of IR services
- Reduced operational burden on ward and daycare services

Structured workflow assessment may also support development of institution-specific fast-track pathways for hepatology patients requiring interventional procedures.

# Conclusion

This audit identified significant workflow inefficiencies affecting therapeutic paracentesis and thoracentesis among hepatology patients.

Retrospective review identified both procedural delays and important limitations in workflow documentation.

Prospective workflow recording enabled accurate measurement of workflow intervals and improved identification of operational bottlenecks.

The findings support development of structured fast-track procedural pathways aimed at improving operational efficiency, multidisciplinary coordination, and patient flow among hepatology patients requiring interventional procedures.
